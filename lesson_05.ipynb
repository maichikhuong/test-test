```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('C:\\Users\\Admin\\Documents\\DataScience\\MindX\\DA\\data\\Customer_Churn.csv')
df
```

```python
col_detect = 'TotalCharges'

df[col_detect] = pd.to_numeric(df[col_detect].replace(" ", ""))
print(df[col_detect].dtype)
df[col_detect] = df[col_detect].fillna(0)
```

```python
col_detect = 'TotalCharges'

Q1 = np.percentile(df[col_detect], 1)
Q1
```

```python
Q3 = np.percentile(df[col_detect], 99)
Q3
```

```python
IQR = Q3 - Q1 
IQR
```

```python
outlier_list = df[ (df[col_detect] < Q1 - 1.5 * IQR) | df[col_detect] > Q3 + 1.5 * IQR ].index

outlier_list
```

```python
df = df.drop(outlier_list, axis=1)
df
```
